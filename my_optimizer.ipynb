```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD, Adagrad
```

```python
# dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[1, 0], [0, 1], [0, 1], [1, 0]])
```

```python
# Create, Compile and Train the model with SGD optimizer
model_sgd = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(2, activation='softmax')
])

sgd_optimizer = SGD(learning_rate=0.01, momentum=0.9)
model_sgd.compile(optimizer=sgd_optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

history_sgd = model_sgd.fit(X, y, epochs=1000, batch_size=4, verbose=0)
```

```python
# Create, Compile and Train the model with Adagrad optimizer
model_adagrad = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(2, activation='softmax')
])

adagrad_optimizer = Adagrad(learning_rate=0.01)
model_adagrad.compile(optimizer=adagrad_optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

history_adagrad = model_adagrad.fit(X, y, epochs=1000, batch_size=4, verbose=0)
```

```python
# Predictions with SGD and Adagrad optimizer
predictions_sgd = np.argmax(model_sgd.predict(X), axis=1)

predictions_adagrad = np.argmax(model_adagrad.predict(X), axis=1)
```

```python
from tensorflow.keras.optimizers import RMSprop

#  Create, Compile, Train and Predict the model with RMSprop optimizer
tahir_model = Sequential([
    Dense(4, activation='relu', input_shape=(2,)),
    Dense(2, activation='softmax')
])

tahir_optimizer = RMSprop(learning_rate=0.001,) and  Adagrad(learning_rate=0.01)
tahir_model.compile(optimizer=tahir_optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

tahir_history = tahir_model.fit(X, y, epochs=1000, batch_size=4, verbose=0)

tahir_predictions = np.argmax(tahir_model.predict(X), axis=1)
```

```python
#plt.figure(figsize=(10, 6))
plt.plot(history_sgd.history['accuracy'], label='SGD')
plt.plot(history_adagrad.history['accuracy'], label='Adagrad')
plt.plot(tahir_history.history['accuracy'], label='Tahir Model')
plt.title('Model Accuracy Comparison')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()
```
